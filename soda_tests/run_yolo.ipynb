## Imports

```python
import os
import json

from PIL import Image

import matplotlib.pyplot as plt

from yolo_detector import ObjectDetectorYOLO 
```

```python
%matplotlib inline
```

## Constants

```python
PREFIX = '../data/SSLAD-2D/labeled/train'
ANNOTATIONS_PREFIX = '../data/SSLAD-2D/labeled/annotations'
```

```python
IOU_THRESHOLD = 0.5
```

```python
CONF_THRESHOLD = 0.4
```

```python
IMAGE_INDEX = 7
```

## Run Predictions

```python
yolo = ObjectDetectorYOLO(weights_path='../weights/yolo11m.pt', 
                          data_path=PREFIX,
                          gt_path=ANNOTATIONS_PREFIX,
                          iou=IOU_THRESHOLD, 
                          conf=CONF_THRESHOLD)
```

```python
files = os.listdir(PREFIX)
```

```python
test_image_path = os.path.join(PREFIX, files[IMAGE_INDEX])
```

```python
result = yolo.predict_one_image(test_image_path)
```

```python
im = Image.open(test_image_path)
```

```python
yolo.draw_pred_bboxes(result, test_image_path)
```

```python
yolo.draw_gt_bboxes(test_image_path)
```
